```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the base URL of the website to scrape
base_url = 'https://sharing-db.club/category/minimal/'

# Loop through the first 8 pages and scrape the tracklists
tracks = []
for page_num in range(1, 9):
    # Construct the URL for the current page
    page_url = f'{base_url}page/{page_num}/'
    print("PAGE URL", page_url)
    
    # Send a GET request to the URL and get the response
    response = requests.get(page_url)

    # Parse the HTML content of the response using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the paragraphs that contain track names
    paragraphs = soup.find_all('p')

    # Extract the track names from the paragraphs
    for p in paragraphs:
        if 'Tracklist' in p.text:
            tracks += [t.strip() for t in p.text.split('\n') if t.strip()]

# Create a dictionary with the track data
track_data = {
    'Track': tracks,
}

# Create a pandas dataframe with the track data
df = pd.DataFrame(track_data)
8
# Split the 'Track' column into separate columns
df[['Track Number', 'Track Name']] = df['Track'].str.split('. ', n=1, expand=True)

# Remove the 'Track' column and reorder the remaining columns
df = df[['Track Number', 'Track Name']]

# Remove the length of time from the track names
df['Track Name'] = df['Track Name'].str.replace(r'\s\([\d:]+\)', '', regex=True)

# Drop track number
df = df.drop('Track Number', axis=1)

# Drop rows with 'None' as the track name
df = df[df['Track Name'] != 'None']

# Reset the index
df = df.reset_index(drop=True)

# Set the max_rows option to None to show all rows
pd.set_option('display.max_rows', None)

# Print the resulting dataframe
print(df)
```
